```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_C_k_meander(k, w, T):
    if k != 0:
        return 2. * np.sin(w * k) / (T * w * k)
    else:
        return 2. / T
```

```python
def plot_meander(n, T):
    x_ = np.linspace(-20, 20, 20000)
    f = np.zeros(20000)
    w = 2. * np.pi / T
    for k in range(-n, n + 1):
        C_k = get_C_k_meander(k, w, T)
        f = f + np.exp(1j * w * k * x_) * C_k
    plt.plot(x_, f)
```

```python
plot_meander(100, 5)
```

```python
T = 5
w = 2. * np.pi / T
coeff = 2 * np.pi / T
C_k = [get_C_k_meander(k, w, T) for k in np.arange(-20, 20, 1)]
plt.plot(np.arange(-20, 20, 1) * coeff, C_k, color = 'blue', linestyle = '-', marker = 'o')
```

```python
plot_meander(100, 10)
```

```python
num_samples = 600
sample_rate = 800
```

```python
def get_sin_func_with_coeff(n):
  x = np.linspace(0.0, num_samples / sample_rate, num_samples, endpoint=False) * n
  f = np.sin(10 * 2 * np.pi * x) + 10 * np.sin(20 * 2 * np.pi * x) + 3 * np.sin(50 * 2 * np.pi * x) + 7 * np.sin(100 * 2 * np.pi * x)
  plt.plot(x, f)
  return f, num_samples, samplerate
```

```python
from scipy.fft import fft, fftfreq

def plot_spectrum(f, num_samples, sample_rate):
  yf = fft(f)
  xf = fftfreq(num_samples, 1 / sample_rate)
  yf_mod = np.abs(yf)
  plt.plot(xf, yf_mod)
```

```python
f, num_samples, samplerate = get_sin_func_with_coeff(1)
```

```python
plot_spectrum(f, num_samples, samplerate)
```

```python
f, num_samples, samplerate = get_sin_func_with_coeff(2)
```

```python
plot_spectrum(f, num_samples, samplerate)
```
