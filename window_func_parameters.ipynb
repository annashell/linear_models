```python
from scipy import signal
import matplotlib.pyplot as plt
from numpy.fft import fft, fftfreq, fftshift
import numpy as np
```

```python
num_samples = 51
```

```python
rect_w = signal.get_window("boxcar", num_samples)

plt.plot(rect_w)
plt.title("Boxcar / rectangular / Dirichlet window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
```

```python
rect_w_spectrum =  np.abs(fft(rect_w, 2048))
frequencies = np.linspace(-0.5, 0.5, len(rect_w_spectrum))
plt.plot(frequencies, fftshift(rect_w_spectrum / abs(rect_w_spectrum).max()))
plt.title("Boxcar / rectangular / Dirichlet window spectrum")
plt.ylabel("Spectral density")
plt.xlabel("Frequencies")
plt.xlim(-0.1,0.1)
plt.grid()
plt.show()
```

```python
A = fft(rect_w, 2048)
freq = np.linspace(-0.5, 0.5, len(A))
response = 20 * np.log10(np.abs(fftshift(A)))
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title("Frequency response of the boxcar window")
plt.ylabel("Normalized magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
plt.ylim(-80,40)
plt.grid()
```
