```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def print_saw_sinewaves(n):
    x_ = np.linspace(0, 20, 20000)
    f = np.zeros(20000)
    for k in range(1, n):
        f_k = (-1)**(k + 1) * 2 * np.sin(k * x_) / k
        f = f + f_k
    plt.grid()
    plt.plot(f)
```

```python
print_saw_sinewaves(2)
```

```python
print_saw_sinewaves(10)
```

```python
print_saw_sinewaves(100)
```

```python
print_saw_sinewaves(1000)
```
