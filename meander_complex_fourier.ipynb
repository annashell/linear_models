```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def print_meander(n, T):
    x_ = np.linspace(-5, 5, 2000)
    f = np.zeros(2000)
    w = 2 * np.pi / T
    for k in range(-n, n + 1):
        if k != 0:
            C_k = 2. * np.sin(w * k) / (T * w * k)
        else:
            C_k = 2. / T    
        f = f + np.exp(1j * w * k * x_) * C_k
    plt.plot(f)
```

```python
print_meander(2, 5)
```

```python
print_meander(5, 5)
```

```python
print_meander(10, 5)
```
